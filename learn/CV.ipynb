# How to measure how bad our model is??

```python
def mse(y_true, y_pred):
    error = y_true - y_pred
    sq_err = np.power(error,2)
    return np.mean(sq_err)

mse(y,y_hat)
```

# How to automate tuning the knob??

\begin{equation}
f(w) = w^2 + 3 * w -5
\end{equation}

```python
# derivative function
def df(w):
    return 2*w+3
```

```python
# initialize weight
w = 10.
```

```python
# gradient descent algorithm
def gd(w):
    w = w - df(w) * 0.01
    return w
for i in range(1000):
    w = gd(w)
w
```

# Linear regression with pytorch

```python
import torch
from torch import tensor,nn
```

```python
n=100
```

```python
x = torch.ones(n,2) 
x[:,0].uniform_(-1.,1)
x[:5]
```

```python
a = tensor([3.,2]); a
```

```python
y = x@a + torch.rand(n)
```

```python
plt.scatter(x[:,0], y);
```

```python
def mse(y_hat, y): return ((y_hat-y)**2).mean()
```

```python
a = tensor([-1.,1])
```

```python
x[:5].shape,a.shape
```

```python
y_hat = x@a
mse(y_hat, y)
```

```python
plt.scatter(x[:,0],y)
plt.scatter(x[:,0],y_hat);
```

```python
a = nn.Parameter(a); a
```

```python
def update():
    y_hat = x@a
    loss = mse(y, y_hat)
    if t % 10 == 0: print(loss)
    loss.backward()
    with torch.no_grad():
        a.sub_(lr * a.grad)
        a.grad.zero_()
```

```python
lr = 1e-1
for t in range(100): update()
```

```python
plt.scatter(x[:,0],y)
y_hat = x@a
plt.scatter(x[:,0],y_hat.detach().numpy());
```
